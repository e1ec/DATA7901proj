```python
import requests
import pandas as pd
```

```python
df = pd.read_csv('dataset_space_API_past_launches_raw.csv')
df
```

```python
falcon_9_df = df[df['BoosterVersion'] == 'Falcon 9']
falcon_9_df
```

```python
missing_percentage = falcon_9_df.isnull().mean() * 100
missing_percentage
```

```python
falcon_9_df['PayloadMass'].fillna(falcon_9_df['PayloadMass'].mean(), inplace=True)

# Check the resulting DataFrame
falcon_9_df
```

```python
def classify_landing_outcome(outcome):
    if outcome in ['True Ocean', 'True RTLS', 'True ASDS']:
        return 1
    else:
        return 0

# Apply the function to the Outcome column to create a new column 'LandingOutcome'
falcon_9_df['LandingOutcome'] = falcon_9_df['Outcome'].apply(classify_landing_outcome)
landing_outcome_counts = falcon_9_df['LandingOutcome'].value_counts()

# Check the resulting DataFrame
print(falcon_9_df.head())
landing_outcome_counts
```

```python
orbit_counts = falcon_9_df['Orbit'].value_counts()

# Calculate the number and occurrence of mission outcome per orbit type
outcome_per_orbit = falcon_9_df.groupby('Orbit')['LandingOutcome'].value_counts().unstack()

orbit_counts, outcome_per_orbit
```

```python
numerical_cols = falcon_9_df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = falcon_9_df.select_dtypes(include=['object']).columns.tolist()

# Calculate the number of launches on each site
launch_site_counts = falcon_9_df['LaunchSite'].value_counts()

numerical_cols, categorical_cols, launch_site_counts
```
